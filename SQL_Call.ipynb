```python
# - Konvertierung von Objekten zu Datentyp Date in der Tabelle Belege
df_Belege['belegdatum']=pd.to_datetime(df_Belege['belegdatum'],format='%Y%m%d', errors='coerce')
df_Belege['erstzulassung']=pd.to_datetime(df_Belege['erstzulassung'],format='%Y%m%d', errors='coerce')
df_Belege['belegerstelltam']=pd.to_datetime(df_Belege['belegerstelltam'],format='%Y%m%d', errors='coerce')
df_Belege['beleggedrucktam']=pd.to_datetime(df_Belege['beleggedrucktam'],format='%Y%m%d', errors='coerce')
df_Belege['fzg_anlage_datum']=pd.to_datetime(df_Belege['fzg_anlage_datum'],format='%Y%m%d', errors='coerce')
```

```python
# - Diese Funktion gibt einen String oder ein Array mit allen Vorkommen zurück und ersetzt die Werte mit dem gewünschten Wert
df_Belege['vk_preis'] = df_Belege['vk_preis'].str.replace(',', '.').astype(float)
df_Belege['mwst_prozent'] = df_Belege['mwst_prozent'].str.replace(',', '.').astype(float)
df_Belege['plz'] = df_Belege['plz'].fillna(0).astype(int).astype(str)
df_Belege['ekpreis'] = df_Belege['ekpreis'].str.replace(',', '.').astype(float)
df_Belege['minuten_pro_aw'] = df_Belege['minuten_pro_aw'].str.replace(',', '.').astype(float)
df_Belege['ktypnr'] = df_Belege['ktypnr'].astype(int)
df_Belege['kmstand'] = df_Belege['kmstand'].astype(int)
df_Belege['menge'] = df_Belege['menge'].str.replace(',', '.').astype(float)
```
